## ✅ How to Download and Install Jupyter Notebook ✅ Step-by-Step Guide

Since direct download from the website isn't available, you can install Jupyter Notebook using one of the following methods:

### Method 1: Using Command Prompt

Pre-requisite: Python must be installed and added to your system's PATH.
Open Command Prompt
Run the following command:

Command Prompt>>pip install notebook

Run Jupyter notebook

### Method 2: Using Anaconda Navigator

Pre-requisite: Anaconda must be installed. You can download it from https://www.anaconda.com


1.Open Anaconda Navigator



2.Go to the Environments tab

3.Select your environment (or create a new one)

4.Search for jupyter in the packages list

5.Check the box next to notebook and click Apply
