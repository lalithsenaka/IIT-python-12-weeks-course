## Escape sequences \n, \r, \t, \b

### New line charactor \n


```python
print("HelloWorld")
```

```python
print("Hello World")
```

```python
print("Hello")
print("World")
```

```python
print("Hello\nWorld")
```

```python
print("Hello\rWorld")
```

```python
print("Today’s rapidly evolving digital \nlandscape, the integration of technology into education has become .")
```

### Charage return \r

```python
print("Today’s rapidly evolving digital \rlandscape, the integration of technology into education has become .")
```

### Tab space \t

```python
print("Today’s rapidly evolving digital \tlandscape, the integration of technology into education has become .")
```

### Back space \b

```python
print("Today’s rapidly evolving digital \blandscape, the integration of technology into education has become .")
```
